```python
import pandas as pd
from geopy.geocoders import Nominatim
import datetime
```

```python
data = pd.read_csv('workshop_survey_data.csv')
data.columns
```

```python
geolocator = Nominatim()
```

```python
data['place_uni']
```

```python
coords = data['place_uni'].apply(geolocator.geocode)
```

```python
data['latlong'] = [[x.latitude, x.longitude] for x in coords]
```

```python
data.to_csv('processed_csv_survey.csv')
```
